# Splitting the dataset into train and test

#Matrix of feature / independent variable

```python
x = dataset.iloc[:,1:-1]
```

#Target variable / dependent variable

```python
y = dataset.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=1234)
```

```python
print(x_train.shape)
print(x_test.shape)
print("----------")
print(y_train.shape)
print(y_test.shape)
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

# Model Building

#Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
classifir_lr = LogisticRegression(random_state=1234)
classifir_lr.fit(x_train,y_train)
```

```python
y_pred_lr = classifir_lr.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,f1_score,precision_score, recall_score
```

```python
acc_lr = accuracy_score(y_test,y_pred_lr)
f1_lr = f1_score(y_test,y_pred_lr)
prec_lr = precision_score(y_test,y_pred_lr)
rec_lr = recall_score(y_test,y_pred_lr)
```

```python
result = pd.DataFrame([['Logistics Regression', acc_lr , f1_lr, prec_lr,rec_lr]],
                      columns=["Model","Accuracy","F1 Score","Precision","Recall"])
```

```python
result 
```

```python
cm = confusion_matrix(y_test,y_pred_lr)
cm
```

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator=classifir_lr,X=x_train,y=y_train,cv=10)
```

```python
print("Accuracy is {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation is {:.2f} %".format(accuracies.std()*100))
```

# Final Model (Logistic Regression)

```python
from sklearn.linear_model import LogisticRegression
classifir = LogisticRegression(C=0.25, class_weight=None, dual=False, fit_intercept=True,
          intercept_scaling=1, max_iter=100, multi_class='warn',
          n_jobs=None, penalty='l2', random_state=1234, solver='liblinear',
          tol=0.0001, verbose=0, warm_start=False)
classifir.fit(x_train,y_train)
```

```python
y_pred_lr_final = classifir.predict(x_test)
```

```python
acc_lr_final = accuracy_score(y_test,y_pred_lr_final)
f1_lr_final = f1_score(y_test,y_pred_lr_final)
prec_lr_final = precision_score(y_test,y_pred_lr_final)
rec_lr_final = recall_score(y_test,y_pred_lr_final)
result_final = pd.DataFrame([['Logistics Regression', acc_lr_final , f1_lr_final, prec_lr_final,rec_lr_final]],
                      columns=["Model","Accuracy","F1 Score","Precision","Recall"])
```

```python
result_final
```

```python
result.append(result_final,ignore_index=True)
```

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator=classifir,X=x_train,y=y_train,cv=10)
```

```python
print("Accuracy is {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation is {:.2f} %".format(accuracies.std()*100))
```
